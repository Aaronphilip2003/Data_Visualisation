```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
games=pd.read_csv('vgsales.csv')
```

```python
games.head()
```

```python
games
```

```python
games_small=games.loc[0:100]
```

```python
games_small
```

```python
fig1=plt.figure(figsize=(20,15))
chart=sns.countplot(x='Year',data=games_small)
chart.set_xticklabels(chart.get_xticklabels(), fontsize=15,rotation=40)
plt.tight_layout()
```

```python
boxfig=plt.figure(figsize=(13,10))
boxchart=sns.boxplot(x='Year',y='Global_Sales',data=games_small)
boxchart.set_xticklabels(boxchart.get_xticklabels(),fontsize=15,rotation=45)
```

```python
games_smaller=games.loc[0:20]
vp=plt.figure(figsize=(13,13))
violinchart=sns.violinplot(x='Year',y='Global_Sales',data=games_smaller,color='red')
violinchart.set_xticklabels(violinchart.get_xticklabels(),fontsize=13,rotation=45,)
```
